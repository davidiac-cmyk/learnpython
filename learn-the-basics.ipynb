## 7. Conditions



```python
# Exercise: grade classifier
score = 86  # example score

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "D or below"

print(f"Score: {score} -> Grade: {grade}")
```

## 8. Loops



```python
# Exercise: sum all even numbers from 1 to 20
total = 0

for n in range(1, 21):
    if n % 2 == 0:
        total += n

print("Sum of even numbers 1–20:", total)

# Simple while loop example
countdown = 5
while countdown > 0:
    print("Countdown:", countdown)
    countdown -= 1
print("Go!")
```

## 9. Functions


```python
# Exercise: define and use a function
def square(x):
    """Return the square of x."""
    return x * x

def hypotenuse(a, b):
    """Return the hypotenuse of a right triangle with legs a and b."""
    return (a ** 2 + b ** 2) ** 0.5

print("square(5) =", square(5))
print("hypotenuse(3, 4) =", hypotenuse(3, 4))
```

## 10. Classes and Objects


```python
# Exercise: simple class for a student
class Student:
    def __init__(self, name, major):
        self.name = name
        self.major = major
        self.credits = 0

    def add_credits(self, amount):
        self.credits += amount

    def summary(self):
        return f"{self.name} – {self.major}, {self.credits} credits"

# Create and use a Student object
s1 = Student("David", "Electrical & Computer Engineering")
s1.add_credits(15)
print(s1.summary())
```

## 11. Dictionaries


```python
# Exercise: use a dictionary to store course grades
grades = {
    "Math": 92,
    "Physics": 88,
    "Programming": 95,
}

print("All grades:", grades)
print("Programming grade:", grades["Programming"])

# Add a new subject
grades["Circuits"] = 90
print("Updated grades:", grades)

# Loop through dictionary items
for course, grade in grades.items():
    print(f"{course}: {grade}")
```
